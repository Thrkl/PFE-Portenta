```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Données d'entraînement

```python
x = [0, 30, 60, 90, 120, 150, 180, 210, 230]
y = [60, 65, 72, 80, 90, 100, 108, 115, 120]
```

```python
def set_Donnees(x, y):
  X_train = np.array(x)
  y_train = np.array(y)
  plt.scatter(x,y)
  plt.show()
  return X_train, y_train
```

```python
X_train, Y_train = set_Donnees(x,y)
```

```python
# Fonction regression
#   arguments : x (nparray : abscisses des données d'entraînement), y (nparray : ordonnées des données d'entraînement)
#             : degre (degre du polynome qui approxime)
#   retourne  : Les coefficients du polynôme (degres décroissants)
def regression (x, y, degree):
  polyfit = np.polyfit(x, y, degree)
  polynome = np.poly1d(polyfit)
  y_pred = polynome(x)
  mse = mean_squared_error(y,y_pred)
  print("Moyenne des carrés des erreurs : ", mse)
  plt.scatter(x, y)
  x_gen = np.linspace(0, 250, 100)
  plt.plot(x_gen, polynome(x_gen))
  plt.show
  return polyfit
```

```python
coeff_poly = regression(X_train, Y_train, 3)
print(coeff_poly)
```

```python
#Tests
x_gen = np.linspace(min(x), max(x), 100)
plt.scatter(X_train, Y_train)
plt.plot(x_gen, 6.00*10+1.17/10*x_gen+1.619/1000*x_gen**2+(-4.36/1000000)*x_gen**3)
```
